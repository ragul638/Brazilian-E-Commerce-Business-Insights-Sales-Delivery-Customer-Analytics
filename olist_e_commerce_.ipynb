```python
import pandas as pd
import numpy as np
```

```python
#to read the file and check weather null values present inside it
df = pd.read_csv('/content/final data for python.csv', encoding='latin1')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
## checking null values
df.isnull().sum().sort_values(ascending=False)
```

```python
# filling blanked values in product category name
df['product_category_name'] = df['product_category_name'].fillna('Unknown')
```

```python
#Create indicator
df['has_review'] = df['review_score'].notnull().astype(int)
```

```python
#fill missing values
df['review_score'] = df['review_score'].fillna(df['review_score'].median())
df
```

```python
df.shape
```

```python
# changing Dtype
date_cols = [
    'order_purchase_timestamp',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], format='%d-%m-%Y', errors='coerce')
```

```python
#changing Dtype
num_cols = [
    'price',
    'freight_value',
    'payment_value',
    'review_score'
]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Changing Dtype
cat_cols = [
    'order_status',
    'payment_type',
    'customer_state',
    'seller_state',
    'product_category_name'
]

for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
# removed duplicate rows
df = df.drop_duplicates()
```

```python
df.shape
```

```python
# creating new column
df['delivery_time'] = (
    df['order_delivered_customer_date'] - df['order_purchase_timestamp']
).dt.days
```

```python
# creating new column
df['delay'] = (
    df['order_delivered_customer_date'] - df['order_estimated_delivery_date']
).dt.days
```

```python
df[['delay', 'review_score']].corr()
```

```python
# creating new column
df['total_order_value'] = df['price'] + df['freight_value']
```

```python
# creating new column
df['order_month'] = df['order_purchase_timestamp'].dt.to_period('M')
```

## Answer Questions

```python
df.groupby('order_month')['total_order_value'].sum().plot()
plt.title("Monthly Sales Trend")
plt.show()
```

**Insight:**
* Sales show a consistent upward/downward trend over time, indicating business growth.
* Any spikes may indicate festive seasons, promotions, or demand surges.

```python
df.groupby('customer_state')['total_order_value'].sum().nlargest(10).plot(kind='bar')
plt.title("Top 10 States by Total Revenue")
plt.show()
```

**Insight:**
* Revenue is heavily concentrated in a few states.
* Business can focus marketing and logistics in high-performing states.

```python
sns.histplot(df['delay'], bins=50)
plt.title("Distribution of Delivery Delayse")
plt.show()
```

**Insight:**
* Most deliveries fall within a normal range, but there are outliers with high delays.
* These delays can negatively impact customer satisfaction and retention.

```python
sns.scatterplot(x=df['delay'], y=df['review_score'])
plt.title("Impact of Delivery Delay on Customer Review Scores")
plt.show()
```

Insight:
* Late deliveries show a negative correlation with review scores,
* indicating logistics delays impact customer satisfaction.

```python
df['payment_type'].value_counts().nlargest(5).plot(kind='pie')
plt.title("Distribution of Payment Methods Used by Customers")
plt.show()
```

**Insight:**
* Indicates customer preference and trust in specific payment types.
* Business can optimize offers or discounts for popular payment methods.

```python
df.groupby('product_category_name')['total_order_value'].sum().nlargest(10).plot(kind='bar')
plt.title("Top 10 Product Categories by Revenue")
plt.show()
```

**Insight:**
* High-performing categories should be prioritized for inventory and marketing.
* Low-performing categories may need strategy revision or removal.

**Key Findings of this:**
- Revenue is concentrated in top states like São Paulo
- Late deliveries negatively impact customer reviews
- Credit card is the most used payment method
- Certain categories have high freight cost relative to price
